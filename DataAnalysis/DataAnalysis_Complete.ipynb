# 6. Data Cleaning

## What is Data Cleaning?
Data cleaning is the process of fixing or removing incorrect, incomplete, corrupted, or irrelevant data.

## Why is it Important?
- **Garbage In, Garbage Out**: Bad data leads to bad results
- Missing values can cause errors in analysis
- Inconsistent data leads to wrong conclusions

## Common Issues:
1. Missing values (NaN, None)
2. Duplicate rows
3. Inconsistent formatting
4. Outliers
5. Wrong data types

```python
# Check for missing values
# WHY: Missing data affects calculations and ML models
print("="*60)
print("üìå Checking for Missing Values")
print("="*60)

# isnull().sum() - Count missing values per column
missing_counts = df.isnull().sum()
print("\nMissing values per column:")
print(missing_counts)

# Total missing values
total_missing = df.isnull().sum().sum()
print(f"\nüìä Total missing values: {total_missing}")
print(f"üìä Total cells: {df.shape[0] * df.shape[1]}")
print(f"üìä Percentage missing: {(total_missing / (df.shape[0] * df.shape[1])) * 100:.2f}%")
```

```python
# Handling missing values
# Create a copy to preserve original data
df_cleaned = df.copy()

# Method 1: dropna() - Remove rows with missing values
# WHY: Simple solution when few missing values, preserves data integrity
# CAUTION: May lose significant data if many missing values
df_dropped = df_cleaned.dropna()
print(f"Original shape: {df.shape}")
print(f"After dropna(): {df_dropped.shape}")
print(f"Rows removed: {df.shape[0] - df_dropped.shape[0]}")
```

```python
# Method 2: fillna() - Fill missing values
# WHY: Preserves row count, uses statistical imputation

# Fill with mean (for numerical columns)
df_filled_mean = df.copy()
df_filled_mean['horsepower'] = df_filled_mean['horsepower'].fillna(df_filled_mean['horsepower'].mean())

# Fill with median (better for skewed data)
df_filled_median = df.copy()
df_filled_median['horsepower'] = df_filled_median['horsepower'].fillna(df_filled_median['horsepower'].median())

print("‚úÖ Missing values handled using mean/median imputation")
print(f"Mean horsepower: {df['horsepower'].mean():.2f}")
print(f"Median horsepower: {df['horsepower'].median():.2f}")
```

# 8. Data Normalization (Min-Max)

## What is Normalization?
Scaling data to a specific range, typically [0, 1].

## Min-Max Normalization Formula:
$$X_{normalized} = \frac{X - X_{min}}{X_{max} - X_{min}}$$

## Why Normalize?
- Makes features comparable on same scale
- Required for many ML algorithms (e.g., neural networks)
- Prevents features with larger ranges from dominating
- Speeds up gradient descent convergence

```python
# Min-Max Normalization
# WHY: Scale values to [0, 1] range for fair comparison

# Normalize 'horsepower' column
df_norm = df.dropna().copy()  # Remove NaN first

# Apply Min-Max formula: (x - min) / (max - min)
hp_min = df_norm['horsepower'].min()
hp_max = df_norm['horsepower'].max()

df_norm['horsepower_normalized'] = (df_norm['horsepower'] - hp_min) / (hp_max - hp_min)

print("="*60)
print("üìå Min-Max Normalization Result")
print("="*60)
print(f"Original range: [{hp_min}, {hp_max}]")
print(f"Normalized range: [{df_norm['horsepower_normalized'].min():.4f}, {df_norm['horsepower_normalized'].max():.4f}]")
print("\nSample comparison:")
print(df_norm[['horsepower', 'horsepower_normalized']].head())
```

# 9. Data Normalization 2 (Z-score)

## Z-score (Standardization) Formula:
$$Z = \frac{X - \mu}{\sigma}$$

Where:
- Œº (mu) = mean of the data
- œÉ (sigma) = standard deviation

## Why Use Z-score?
- Centers data around 0
- Measures how many standard deviations from mean
- Useful for identifying outliers (|Z| > 3)
- Preferred when data follows normal distribution

```python
# Z-score Normalization (Standardization)
# WHY: Centers data at 0, measures distance in standard deviations

# Apply Z-score formula: (x - mean) / std
hp_mean = df_norm['horsepower'].mean()
hp_std = df_norm['horsepower'].std()

df_norm['horsepower_zscore'] = (df_norm['horsepower'] - hp_mean) / hp_std

print("="*60)
print("üìå Z-score Normalization Result")
print("="*60)
print(f"Mean: {hp_mean:.2f}")
print(f"Standard Deviation: {hp_std:.2f}")
print(f"Z-score range: [{df_norm['horsepower_zscore'].min():.4f}, {df_norm['horsepower_zscore'].max():.4f}]")
print("\nSample comparison:")
print(df_norm[['horsepower', 'horsepower_normalized', 'horsepower_zscore']].head())
```

# 10. Binning

## What is Binning?
Converting continuous numeric data into discrete categories (bins).

## Why Use Binning?
- Simplifies complex data
- Reduces noise and outlier effects
- Creates categorical features for analysis
- Enables grouping for visualization

```python
# Binning - Converting numeric to categorical
# WHY: Group continuous values into meaningful categories

df_binned = df.dropna().copy()

# Create bins for horsepower
bins = [0, 75, 125, 175, 250]  # Define bin edges
labels = ['Low', 'Medium', 'High', 'Very High']  # Category names

# pd.cut() - Bin based on value ranges
df_binned['hp_category'] = pd.cut(df_binned['horsepower'], bins=bins, labels=labels)

print("="*60)
print("üìå Binning Result")
print("="*60)
print("\nBin ranges:")
for i in range(len(labels)):
    print(f"  {labels[i]}: {bins[i]} - {bins[i+1]}")

print("\nCategory distribution:")
print(df_binned['hp_category'].value_counts())

print("\nSample data:")
print(df_binned[['horsepower', 'hp_category']].head(10))
```

# 11. Binning 2 (Equal Frequency)

## Equal Frequency Binning (Quantile-based)
Divides data so each bin has approximately the same number of observations.

## Difference from Fixed-width Binning:
- **Fixed-width (cut)**: Equal range sizes, unequal counts
- **Equal-frequency (qcut)**: Equal counts, variable range sizes

## When to Use:
- When you want balanced groups
- For creating percentile-based categories

```python
# Equal Frequency Binning using qcut()
# WHY: Creates bins with equal number of samples

# qcut - Quantile-based binning (equal frequency)
df_binned['hp_quantile'] = pd.qcut(df_binned['horsepower'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

print("="*60)
print("üìå Equal Frequency Binning (Quartiles)")
print("="*60)
print("\nEach quartile has approximately equal samples:")
print(df_binned['hp_quantile'].value_counts().sort_index())

# Show the ranges for each quantile
print("\nQuantile ranges:")
for q in ['Q1', 'Q2', 'Q3', 'Q4']:
    subset = df_binned[df_binned['hp_quantile'] == q]['horsepower']
    print(f"  {q}: {subset.min():.0f} - {subset.max():.0f}")
```

3. **Combine**: Merge results

## Why Use GroupBy?
- Calculate statistics per category
- Compare groups
- Aggregate data for reporting

```python
# GroupBy - Split-Apply-Combine
# WHY: Calculate statistics for each group

print("="*60)
print("üìå GroupBy Operations")
print("="*60)

# Group by cylinders and calculate mean
print("\nüîß Average MPG by Cylinders:")
print(df.groupby('cylinders')['mpg'].mean().round(2))

# Multiple aggregations
print("\nüìä Multiple Statistics by Cylinders:")
print(df.groupby('cylinders')['mpg'].agg(['mean', 'std', 'min', 'max']).round(2))

# Group by multiple columns
print("\nüìÖ Average MPG by Cylinders and Model Year:")
print(df.groupby(['cylinders', 'model-year'])['mpg'].mean().unstack().round(1).head())
```

# 18. ANOVA (Analysis of Variance)

## What is ANOVA?
Statistical test to compare means across multiple groups.

## When to Use:
- Compare 3+ group means
- Determine if differences are statistically significant
- Example: Does MPG differ significantly across cylinder counts?

## Key Outputs:
- F-statistic: Ratio of between-group to within-group variance
- p-value: Probability that differences occurred by chance
- p < 0.05 ‚Üí Statistically significant difference

```python
# ANOVA - Compare means across groups
# WHY: Test if MPG differs significantly by cylinder count

from scipy import stats

# Prepare groups - MPG for each cylinder count
df_anova = df.dropna()
groups = [group['mpg'].values for name, group in df_anova.groupby('cylinders')]

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(*groups)

print("="*60)
print("üìå ANOVA: MPG by Cylinders")
print("="*60)
print(f"\nF-statistic: {f_statistic:.4f}")
print(f"p-value: {p_value:.2e}")

if p_value < 0.05:
    print("\n‚úÖ Result: SIGNIFICANT (p < 0.05)")
    print("   MPG differs significantly across cylinder groups")
else:
    print("\n‚ùå Result: NOT SIGNIFICANT (p >= 0.05)")
    print("   No significant difference in MPG across groups")

# Group means for reference
print("\nüìä Mean MPG by Cylinders:")
print(df_anova.groupby('cylinders')['mpg'].mean().round(2))
```

# 20. Correlation 2 - Pearson Correlation

## Pearson Correlation Coefficient
Most common correlation measure for linear relationships.

## Formula:
$$r = \frac{\sum(X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum(X_i - \bar{X})^2}\sqrt{\sum(Y_i - \bar{Y})^2}}$$

## Outputs:
- **Correlation coefficient (r)**: Strength and direction
- **p-value**: Statistical significance

```python
# Pearson Correlation with p-value
# WHY: Get both correlation strength and statistical significance

from scipy.stats import pearsonr

df_corr = df.dropna()

# Calculate Pearson correlation between weight and MPG
corr, p_value = pearsonr(df_corr['weight'], df_corr['mpg'])

print("="*60)
print("üìå Pearson Correlation: Weight vs MPG")
print("="*60)
print(f"\nCorrelation coefficient (r): {corr:.4f}")
print(f"p-value: {p_value:.2e}")

# Interpret
strength = "Strong" if abs(corr) > 0.7 else "Moderate" if abs(corr) > 0.4 else "Weak"
direction = "negative" if corr < 0 else "positive"
print(f"\nüìä Interpretation: {strength} {direction} correlation")
print("   Heavier cars tend to have lower MPG")
```

# 21. Simple Linear Regression (Theory)

## What is Linear Regression?
A method to model the relationship between a dependent variable (y) and one independent variable (x).

## Formula:
$$y = b_0 + b_1x$$

Where:
- y: Dependent variable (what we predict)
- x: Independent variable (predictor)
- b‚ÇÄ: Intercept (y-value when x=0)
- b‚ÇÅ: Slope (change in y for unit change in x)

## Assumptions:
1. Linear relationship
2. Independence of errors
3. Homoscedasticity (constant variance)
4. Normality of residuals

# 22. Multiple Linear Regression (Theory)

## Multiple Linear Regression
Extends simple regression to multiple predictor variables.

## Formula:
$$y = b_0 + b_1x_1 + b_2x_2 + ... + b_nx_n$$

## Example for Car MPG:
$$MPG = b_0 + b_1 \times horsepower + b_2 \times weight + b_3 \times acceleration$$

## Advantages:
- Captures multiple influences
- More realistic models
- Better predictions

# 23. Simple Linear Regression using Python

## Implementation
Using scikit-learn's LinearRegression class.

```python
# Simple Linear Regression - One predictor
# WHY: Model relationship between weight and MPG

from sklearn.linear_model import LinearRegression

df_reg = df.dropna()

# Prepare data
# X must be 2D: reshape(-1, 1) converts 1D array to 2D column
X = df_reg[['weight']].values  # Independent variable (2D)
y = df_reg['mpg'].values       # Dependent variable (1D)

# Create and fit model
lm = LinearRegression()
lm.fit(X, y)

# Get predictions
y_pred = lm.predict(X)

print("="*60)
print("üìå Simple Linear Regression: Weight ‚Üí MPG")
print("="*60)
print(f"\nIntercept (b‚ÇÄ): {lm.intercept_:.4f}")
print(f"Slope (b‚ÇÅ): {lm.coef_[0]:.6f}")
print(f"\nEquation: MPG = {lm.intercept_:.2f} + ({lm.coef_[0]:.6f}) √ó Weight")
print(f"\nInterpretation: For every 1 lb increase in weight,")
print(f"                MPG decreases by {abs(lm.coef_[0]):.4f}")
```

```python
# Multiple Linear Regression - Multiple predictors
# WHY: Use multiple features for better predictions

# Prepare multiple features
X_multi = df_reg[['horsepower', 'weight', 'acceleration']].values
y = df_reg['mpg'].values

# Create and fit model
lm_multi = LinearRegression()
lm_multi.fit(X_multi, y)

# Get predictions
y_pred_multi = lm_multi.predict(X_multi)

print("="*60)
print("üìå Multiple Linear Regression")
print("="*60)
print(f"\nIntercept (b‚ÇÄ): {lm_multi.intercept_:.4f}")
print("\nCoefficients:")
features = ['horsepower', 'weight', 'acceleration']
for i, feat in enumerate(features):
    print(f"  {feat}: {lm_multi.coef_[i]:.6f}")

print(f"\nEquation:")
print(f"  MPG = {lm_multi.intercept_:.2f}", end='')
for i, feat in enumerate(features):
    sign = '+' if lm_multi.coef_[i] >= 0 else '-'
    print(f" {sign} {abs(lm_multi.coef_[i]):.6f}√ó{feat}", end='')
print()
```

# 32. Mean Squared Error (MSE)

## What is MSE?
Average of squared differences between actual and predicted values.

## Formula:
$$MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$$

## Why MSE?
- Penalizes larger errors more heavily
- Always positive
- Same units as y¬≤ (use RMSE for original units)

```python
# Mean Squared Error
# WHY: Measure prediction error magnitude

from sklearn.metrics import mean_squared_error, mean_absolute_error

print("="*60)
print("üìå Error Metrics")
print("="*60)

# Simple regression errors
mse_simple = mean_squared_error(y, y_pred)
rmse_simple = np.sqrt(mse_simple)

print("\nSimple Regression:")
print(f"  MSE  = {mse_simple:.4f}")
print(f"  RMSE = {rmse_simple:.4f} (same units as MPG)")

# Multiple regression errors
mse_multi = mean_squared_error(y, y_pred_multi)
rmse_multi = np.sqrt(mse_multi)

print("\nMultiple Regression:")
print(f"  MSE  = {mse_multi:.4f}")
print(f"  RMSE = {rmse_multi:.4f}")

print("\nüìä Lower MSE/RMSE = Better predictions")
print(f"üìä Multiple regression reduced RMSE by {((rmse_simple - rmse_multi)/rmse_simple)*100:.1f}%")
```

# 33. R Squared (R¬≤) ‚Äì Concept

## What is R¬≤?
Coefficient of Determination - measures how much variance in y is explained by the model.

## Formula:
$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{\sum(y_i - \hat{y}_i)^2}{\sum(y_i - \bar{y})^2}$$

## Interpretation:
- **R¬≤ = 1**: Perfect fit (100% variance explained)
- **R¬≤ = 0**: No explanatory power
- **R¬≤ = 0.80**: 80% of variance explained

## Limitations:
- Always increases with more features
- Use Adjusted R¬≤ for multiple regression

# 34. R Squared 2 ‚Äì With Example

```python
# R¬≤ Analysis and Comparison
# WHY: Comprehensive understanding of model performance

from sklearn.metrics import r2_score

print("="*60)
print("üìå R¬≤ Score Comparison")
print("="*60)

# Calculate R¬≤ for all models
r2_results = {
    'Simple (weight)': r2_score(y, y_pred),
    'Multiple (3 features)': r2_score(y, y_pred_multi),
    'Polynomial (degree 2)': r2_score(y_poly, poly_model.predict(X_poly_prep))
}

print("\nüìä Model Performance Comparison:")
print("-" * 40)
for model, r2 in r2_results.items():
    bar = '‚ñà' * int(r2 * 40)
    print(f"{model:25} | R¬≤ = {r2:.4f} | {bar}")

# Best model interpretation
best_model = max(r2_results, key=r2_results.get)
best_r2 = r2_results[best_model]

print(f"\n‚úÖ Best Model: {best_model}")
print(f"   R¬≤ = {best_r2:.4f}")
print(f"   Interpretation: {best_r2*100:.1f}% of variance in MPG is explained by this model")

# What R¬≤ values mean in practice
print("\nüìñ R¬≤ Guidelines:")
print("   R¬≤ > 0.9  ‚Üí Excellent fit")
print("   R¬≤ > 0.7  ‚Üí Good fit")
print("   R¬≤ > 0.5  ‚Üí Moderate fit")
print("   R¬≤ < 0.5  ‚Üí Poor fit")
```
