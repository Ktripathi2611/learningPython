```python
import tkinter as tk
import tkinter.font as tkFont

# Create main window
window = tk.Tk()
window.title("Entry Component")
window.minsize(width=600, height=300)
window.maxsize(width=700, height=400)
window.config(padx=20, pady=20)

# Custom font (optional)
custom_font = tkFont.Font(family="Helvetica", size=14, weight="bold")

# ADD THE LABEL
label = tk.Label(window, text="This is entry component", font=custom_font)
label.pack(pady=10)

# ADD THE ENTRY with placeholder behavior
def on_entry_click(event):
    if entry.get() == "Enter your name":
        entry.delete(0, "end")  # clear placeholder
        entry.config(fg="black")

def on_focusout(event):
    if entry.get() == "":
        entry.insert(0, "Enter your name")
        entry.config(fg="grey")

entry = tk.Entry(window, fg="grey")
entry.insert(0, "Enter your name")
entry.bind("<FocusIn>", on_entry_click)
entry.bind("<FocusOut>", on_focusout)
entry.pack(padx=10, pady=10)

# ADD THE BUTTON
def button_click():
    label.config(text=f"You entered: {entry.get()}")

button = tk.Button(window, text="Submit", command=button_click)
button.pack(pady=10)

# START THE MAIN LOOP
window.mainloop()
```

```python
import tkinter as tk
from tkinter import font

# Create main window
window = tk.Tk()
window.title("Login Form")
window.minsize(width=400, height=300)

# Custom font
custom_font = font.Font(family="Helvetica", size=12, weight="bold")

# Username label and entry
username_label = tk.Label(window, text="Enter your username:", font=custom_font)
username_label.pack(pady=5)
username_entry = tk.Entry(window, width=30)
username_entry.pack(pady=5)

# Password label and entry
password_label = tk.Label(window, text="Enter your password:", font=custom_font)
password_label.pack(pady=5)
password_entry = tk.Entry(window, width=30, show="*")
password_entry.pack(pady=5)

# Email label and entry
email_label = tk.Label(window, text="Enter your email:", font=custom_font)
email_label.pack(pady=5)
email_entry = tk.Entry(window, width=30)
email_entry.pack(pady=5)

# Remember me checkbox
remember_var = tk.IntVar()
remember_check = tk.Checkbutton(window, text="Remember me", variable=remember_var)
remember_check.pack(pady=5)

# Login button
def login_action():
    username = username_entry.get()
    password = password_entry.get()
    email = email_entry.get()
    print(f"Username: {username}, Password: {password}, Email: {email}")

login_button = tk.Button(window, text="Login", font=custom_font, command=login_action)
login_button.pack(pady=10)

# Exit button
exit_button = tk.Button(window, text="Exit", command=window.quit)
exit_button.pack(pady=5)

# Run the application
window.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

# Main window
window = tk.Tk()
window.title("Tkinter Demo Project")
window.minsize(width=600, height=600)

# --- 2. Label, Title, minsize, pack ---
label = tk.Label(window, text="Welcome to Tkinter GUI Demo!", font=("Arial", 16))
label.pack(pady=10)

# --- 3. Button and change label text ---
def change_text():
    label.config(text="You clicked the button!")

btn = tk.Button(window, text="Click Me", command=change_text)
btn.pack(pady=5)

# --- 4. Entry component ---
entry_label = tk.Label(window, text="Enter your name:")
entry_label.pack()
entry = tk.Entry(window, width=30)
entry.pack(pady=5)

# --- 5. ttk example ---
ttk_label = ttk.Label(window, text="This is a ttk Label (modern look).")
ttk_label.pack(pady=5)

# --- 6. Quit button ---
quit_btn = ttk.Button(window, text="Quit", command=window.quit)
quit_btn.pack(pady=5)

# --- 7. Frames and background colors ---
frame = tk.Frame(window, bg="lightblue", bd=2, relief="sunken")
frame.pack(pady=10, fill="x")
frame_label = tk.Label(frame, text="Inside a Frame", bg="lightblue")
frame_label.pack(pady=5)

# --- 8. Text, Separator, and padding ---
text_widget = tk.Text(window, height=4, width=40)
text_widget.insert(tk.END, "This is a Text widget.\nYou can type multiple lines here.")
text_widget.pack(pady=5)

separator = ttk.Separator(window, orient="horizontal")
separator.pack(fill="x", pady=10)

# --- 9. Checkbutton widget ---
check_var = tk.IntVar()
check_btn = tk.Checkbutton(window, text="I agree to terms", variable=check_var)
check_btn.pack(pady=5)

# --- 10. Radiobutton widget ---
radio_var = tk.StringVar(value="Option1")
radio1 = tk.Radiobutton(window, text="Option 1", variable=radio_var, value="Option1")
radio2 = tk.Radiobutton(window, text="Option 2", variable=radio_var, value="Option2")
radio1.pack()
radio2.pack()

# --- 11. Listbox widget ---
listbox = tk.Listbox(window, height=5)
items = ["Python", "Java", "C++", "JavaScript", "Go"]
for item in items:
    listbox.insert(tk.END, item)
listbox.pack(pady=5)

# --- 12. Spinbox widget ---
spinbox = tk.Spinbox(window, from_=1, to=10, width=5)
spinbox.pack(pady=5)

# Run the application
window.mainloop()
```
