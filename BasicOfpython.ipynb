## 6. Primitive Data Types

Python has several built-in data types for storing different kinds of values.

### Integer (int)

Whole numbers without decimal points.

```python
age=25
print(age)
print(type (age))
```

### Float

Numbers with decimal points.

```python
age=25.5
print(age)
print(type (age))
```

### String (str)

Text data enclosed in quotes.

```python
# string
name="sachin"
```

```python

print(name)
print(type (name))
```

### Boolean (bool)

True or False values.

```python
# boolean
# True
# False
```

```python
name=True
print(name)
print(type (name))
```

### NoneType

Represents the absence of a value.

```python
# specialdatatype
name=None
```

```python
name
print(name)
print(type (name))
```
