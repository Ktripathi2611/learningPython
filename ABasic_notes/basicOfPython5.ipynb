# ðŸŽ› Module 8: Control Statements in Python

Control statements allow us to **control the flow of execution** in a program.

---

## ðŸ” For Loop
Used to iterate over a sequence (list, tuple, string, dictionary, etc.).

```python
# Example: Iterating over a list
for num in [1,2,3,4,5]:
    print(num)
```

### Output:
```
1
2
3
4
5
```

---

## ðŸŸ© I. The `if` Block
Executes a block of code if the condition is **True**.

```python
x = 10
if x > 5:
    print("x is greater than 5")
```

---

## ðŸŸ¨ 2. `if-else` Block
Provides an alternative block if the condition is **False**.

```python
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

---

## ðŸŸ¦ 3. `if-elif-else` Block
Checks multiple conditions in sequence.

```python
x = 10
if x < 5:
    print("x is less than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is greater than 5")
```

---

## ðŸŸª 4. Nested `if` Statements
An `if` inside another `if`.

```python
x = 15
if x > 10:
    if x % 2 == 0:
        print("x is greater than 10 and even")
    else:
        print("x is greater than 10 and odd")
```

---

## ðŸŸ¥ 5. The Ternary Operator
A shorthand for `if-else` in one line.

```python
x = 7
result = "Even" if x % 2 == 0 else "Odd"
print(result)
```

### Output:
```
Odd
```
