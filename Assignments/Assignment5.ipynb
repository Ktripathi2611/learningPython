Task 1: Create a Dictionary of Student Marks

Problem Statement: Write a Python program that:
1.   Creates a dictionary where student names are keys and their marks are values.
2.   Asks the user to input a student's name.
3.   Retrieves and displays the corresponding marks.
4.   If the studentâ€™s name is not found, display an appropriate message.

```python
# Create a dictionary called student_marks where each key is a student's name
# and the value is their marks (integer).
student_marks = {
    "Alice": 85,
    "Bob": 78,
    "Charlie": 92,
    "David": 67,
    "Eva": 90
}

# Prompt the user to enter a student's name.
# The input() function takes user input as a string and stores it in the variable 'name'.
name = input("Enter the student's name: ")

# Check if the entered name exists as a key in the student_marks dictionary.
if name in student_marks:
    # If the name exists, print the student's marks using f-string formatting.
    # f"{name}'s marks: {student_marks[name]}" dynamically inserts the student's name and marks.
    print(f"{name}'s marks: {student_marks[name]}")
else:
    # If the name does not exist in the dictionary, print a message saying student not found.
    print("Student not found in the dictionary.")
```

Task 2: Demonstrate List Slicing 
Problem Statement: Write a Python program that:
1.   Creates a list of numbers from 1 to 10.
2.   Extracts the first five elements from the list.
3.   Reverses these extracted elements.
4.   Prints both the extracted list and the reversed list


```python
# Create a list of numbers from 1 to 10 using range().
# range(1, 11) generates numbers from 1 up to (but not including) 11.
numbers = list(range(1, 11))

# Extract the first five elements from the list using slicing.
# numbers[:5] means "take elements from index 0 up to (but not including) index 5".
first_five = numbers[:5]

# Reverse the extracted list using slicing with a step of -1.
# first_five[::-1] means "take the list from end to start".
reversed_list = first_five[::-1]

# Print the extracted list (first five numbers).
print("Extracted list:", first_five)

# Print the reversed version of the extracted list.
print("Reversed list:", reversed_list)
```
